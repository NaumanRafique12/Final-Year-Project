```python
import cv2
```

```python
import os
```

```python
videoPath = "./plain.mp4"
```

```python
dir = "Data"+videoPath.split(".")[-2]+"/"
if not os.path.exists(dir):
    os.makedirs(dir)
```

```python
vidcap = cv2.VideoCapture(videoPath)
success,image = vidcap.read()
count = 0
success = True
while count < 1500:
    cv2.imwrite(dir+"frame%d.jpg" % count, image)     # save frame as JPEG file
    success,image = vidcap.read()
    print ("Reading Frame ",count)
    count += 1
```

```python
vidcap.release()
```
